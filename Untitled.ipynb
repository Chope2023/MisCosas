````python
# Crear una función para calcular estadísticas descriptivas y devolverlas en un DataFrame
def calculate_boxplot_stats(data, target, features):
    stats = []
    for feature in features:
        df_churn = data[data[target] == 1][feature].describe()
        df_no_churn = data[data[target] == 0][feature].describe()

        stats.append({
            'Feature': feature,
            'Churn Median': df_churn['50%'],
            'Churn Q1': df_churn['25%'],
            'Churn Q3': df_churn['75%'],
            'No Churn Median': df_no_churn['50%'],
            'No Churn Q1': df_no_churn['25%'],
            'No Churn Q3': df_no_churn['75%'],
            'Churn Mean': df_churn['mean'],
            'No Churn Mean': df_no_churn['mean'],
        })

    return pd.DataFrame(stats)

# Lista de variables continuas
continuous_vars = [
    'Account length', 'Area code', 'Number vmail messages',
    'Total day minutes', 'Total day calls', 'Total day charge',
    'Total eve minutes', 'Total eve calls', 'Total eve charge',
    'Total night minutes', 'Total night calls', 'Total night charge',
    'Total intl minutes', 'Total intl calls', 'Total intl charge',
    'Customer service calls'
]

# Calcular estadísticas descriptivas para las variables continuas
stats_df = calculate_boxplot_stats(df, 'Churn', continuous_vars)

# Mostrar las tablas
print(stats_df)
````

Análisis de Componentes Principales (PCA)
---

El Análisis de Componentes Principales (PCA) es una técnica que ayuda a reducir el número de variables en un conjunto de datos mientras mantiene la mayor cantidad posible de la información original.

En simple, primero, transforma los datos en una nueva matriz donde cada nueva variable, llamada componente principal, es una combinación lineal de las variables originales. Estas componentes se ordenan según la cantidad de varianza o información que conservan.

Para seleccionar las componentes principales más útiles, se calcula cuánta información retienen en conjunto. Se eligen aquellas que juntas explican una alta proporción de la varianza total. Esto permite reducir la dimensionalidad del conjunto de datos sin perder mucha información importante.

Aunque PCA puede ser muy útil en conjuntos de datos con muchas variables, en problemas más simples puede que no sea necesario. La técnica también ayuda a reducir el ruido y mejorar la eficiencia computacional al simplificar el modelo

En este caso, por la difucltad del libro, no se hace necesario este tipo de tecnicas de reduccion

Codificación de Variables Categóricas
---
Antes de normalizar las variables, es fundamental codificar las variables categóricas, es decir, convertir las categorías en valores numéricos. Este paso es crucial porque muchos algoritmos de machine learning requieren que todas las entradas sean numéricas.

Es importante diferenciar entre variables categóricas que tienen un orden intrínseco y aquellas que no. Por ejemplo, si las categorías son "Bueno", "Aceptable" y "Malo", podemos asignarles números como 1, 2 y 3. Esto sugiere una relación ordinal, donde el número refleja un grado de calidad. Sin embargo, en el caso de variables como "Estado", que representan identidades sin un orden implícito, utilizaremos técnicas que evitan asignar un peso implícito a las categorías.

Las técnicas más comunes son:

Codificación de Etiquetas (Label Encoding):

Convierte cada categoría en un número entero. Es adecuada para variables ordinales donde el orden tiene significado. Por ejemplo, "Rojo" puede ser 0, "Verde" 1, y "Azul" 2.
Codificación One-Hot (One-Hot Encoding):

Crea una columna binaria para cada categoría, asignando un 1 en la columna correspondiente a la categoría de un registro y 0 en las demás. Es ideal para variables nominales, como los "Estados", donde el número no debe implicar un orden o peso. Para este caso, se utilizará One-Hot Encoding, ya que sólo necesitamos una representación numérica sin asignar un peso a cada estado.

````python
# Realizar el One-Hot Encoding
df_encoded = pd.get_dummies(df_filtered, columns=['State'], prefix='State')

# Mostrar las primeras filas del DataFrame actualizado
print(df_encoded.head())
df_encoded.dtypes
````

Reduccion de dimensionalidad PCA
---

Uno de los desafíos que presentan los conjuntos de datos es que, a veces, existen variables independientes que están fuertemente correlacionadas, es decir, que dicen lo mismo pero de una manera diferente.

Por ejemplo, tenemos una variable llamada "Cargo" y otra llamada "Salario". Es evidente que, entre mayor sea el cargo, mayor será el salario, entonces cargaríamos el análisis con una variable de algún modo duplicada.

# **PCA**
NOTA : es importante destacar que el analisis por PCA se ejecuta despúes de haber balanceado la variable dependiente y no antes, ya que esto arrastra el sezgo por cantidad de samples

Para la detección de este tipo de correlaciones se utiliza el método de análisis de componentes principales (PCA, por sus siglas en inglés), donde la idea es reducir la dimensionalidad del conjunto de datos mientras se retiene la mayor cantidad posible de variabilidad presente en los datos originales.

El PCA transforma las variables originales en un nuevo conjunto de variables, llamadas componentes principales, que son ortogonales entre sí.

Esto es importante de retener: es un nuevo conjunto de variables donde cada variable se llamará Componente Principal. Así, en este caso, tenemos 68 variables de entrada, por lo que se generará una matriz de igual dimensión pero con nuevas variables denominadas Componentes Principales. Estas nuevas variables son combinaciones lineales de las variables originales y capturan la mayor parte de la variabilidad en los datos con menos dimensiones. La reducción dimensional a través de PCA no solo simplifica el modelo al reducir el número de variables, sino que también puede ayudar a mejorar la interpretación y el rendimiento del modelo al eliminar redundancias y reducir el ruido en los dato

Estas componentes principales se ordenan de tal manera que la primera componente captura la mayor parte de la variabilidad en los datos, la segunda componente captura la mayor parte de la variabilidad restante, y así sucesivamente.


Las componentes principales pueden ser representadas gráficamente en un espacio de hasta tres dimensiones (x, y, z) si se seleccionan las primeras tres componentes principales.

La reducción de dimensionalidad tiene varios beneficios:

Reducción del ruido: Al eliminar variables redundantes, se puede reducir el ruido en los datos, lo que puede mejorar el rendimiento de los modelos de machine learning.

Mejora de la interpretabilidad: Con menos variables, es más fácil interpretar los resultados y entender qué factores son los más importantes.

Reducción del tiempo de procesamiento: Menos variables significan menos datos que procesar, lo que puede reducir significativamente el tiempo de entrenamiento de los modelos.

Prevención del sobreajuste: Con menos variables, hay menos riesgo de que el modelo se ajuste demasiado a las particularidades del conjunto de entrenamiento y no generalice bien a datos nuevos.

Sin embargo, es importante tener en cuenta que la reducción de dimensionalidad puede resultar en la pérdida de información, por lo que es crucial encontrar un equilibrio adecuado entre simplificación y retención de variabilidad.

````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Separar características y variable objetivo
X_balanced_robust = df_Smote.drop('Churn', axis=1)
y_balanced_robust = df_Smote['Churn']

# Aplicar PCA
pca = PCA()
X_pca_robust = pca.fit_transform(X_balanced_robust)

# Obtener la varianza explicada y acumulada
explained_variance_ratio_robust = pca.explained_variance_ratio_
cumulative_variance_robust = explained_variance_ratio_robust.cumsum()


# Gráfico de la varianza explicada acumulada
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance_ratio_robust) + 1), cumulative_variance_robust, marker='o', linestyle='--', color='b')
plt.title('Varianza Explicada por PC PARA DF ROBUST', fontsize=24)
plt.xlabel('Número de Componentes Principales', fontsize=18)
plt.ylabel('Varianza Explicada Acumulada', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)
plt.show()

# Gráfico de la varianza explicada por cada componente principal
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance_ratio_robust) + 1), explained_variance_ratio_robust, marker='o', linestyle='--', color='r')
plt.title('Varianza Explicada por PC PARA DF ROBUST', fontsize=24)
plt.xlabel('Número de Componentes Principales', fontsize=18)
plt.ylabel('Varianza Explicada', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)
plt.show()
````

Al observar las varianzas explicadas acumulativas de los componentes principales, notamos que los tres primeros componentes explican el 99.9% de la variabilidad de los datos. Esto significa que nuestro conjunto de datos, que inicialmente tenía 8 variables, puede ser reducido a solo 3 componentes principales sin perder casi ninguna información relevante.

Este proceso se denomina reducción de dimensionalidad, ya que permite simplificar el conjunto de datos al reducir el número de variables necesarias para los algoritmos de Machine Learning. En nuestro caso, los datos ahora están representados únicamente por la variable Churn y las tres componentes principales, lo que facilita la aplicación de los algoritmos de ML.

Bias y Varianza
---

Los conceptos de bias (sesgo) y varianza son fundamentales para entender cómo los modelos de Machine Learning se ajustan a los datos.

Bias (Sesgo):

Modelos Estadísticos Formales: Los modelos basados en cálculos estadísticos, como la regresión lineal, suelen ser más rígidos y simplificados. Esto puede llevar a un alto sesgo porque estos modelos asumen ciertas relaciones estructurales entre las variables que pueden no capturar la complejidad real de los datos. En otras palabras, un alto sesgo puede hacer que el modelo subestime la variabilidad en los datos, llevándolo a un ajuste deficiente.


Redes Neuronales: Las redes neuronales, por otro lado, son más flexibles debido a su capacidad para aprender relaciones complejas y no lineales entre las variables. Esto les permite ajustar mejor los datos, reduciendo el sesgo. Sin embargo, esta flexibilidad puede tener sus propios desafíos, como la necesidad de grandes cantidades de datos para evitar el sobreajuste.


```


```

Varianza:

Modelos Estadísticos Formales: Tienen generalmente una menor varianza, ya que su estructura rígida limita la capacidad del modelo para adaptarse a pequeñas fluctuaciones en los datos de entrenamiento. Esto puede ser beneficioso para evitar el sobreajuste, pero también puede limitar la capacidad del modelo para capturar patrones complejos.


Redes Neuronales: Tienen una mayor varianza debido a su capacidad para adaptarse y modelar datos complejos. Esto significa que pueden ajustarse demasiado a los datos de entrenamiento, lo que puede llevar a sobreajuste si no se maneja correctamente.


En resumen, los modelos estadísticos formales tienden a tener un mayor sesgo y menor varianza, mientras que las redes neuronales pueden tener un menor sesgo pero mayor varianza. La clave es encontrar un equilibrio adecuado entre sesgo y varianza para mejorar el rendimiento del modelo en datos no vistos.

Optimizacion de parametros
---
Cada algoritmo de Machine Learning tiene un conjunto de parámetros, conocidos como hiperparámetros, que controlan el comportamiento del modelo durante el entrenamiento. Estos hiperparámetros son fundamentales para ajustar la función de predicción y lograr el mejor rendimiento posible del modelo.

Hiperparámetros:

Los hiperparámetros son configuraciones que no se aprenden directamente durante el entrenamiento del modelo, sino que deben ser establecidos antes de este proceso. Ejemplos incluyen la tasa de aprendizaje en redes neuronales, el número de árboles en un Random Forest, o el valor de k en K-Nearest Neighbors (KNN).
Optimización de Hiperparámetros:

Búsqueda en Rejilla (Grid Search): Esta técnica implica definir un rango de valores posibles para cada hiperparámetro y luego entrenar el modelo con todas las combinaciones posibles de estos valores. Esto permite encontrar la combinación que ofrece el mejor rendimiento según un criterio definido, como precisión (precision), recuperación (recall) o puntuación F1.
Búsqueda Aleatoria (Random Search): En lugar de probar todas las combinaciones posibles, esta técnica selecciona aleatoriamente combinaciones de hiperparámetros dentro de los rangos especificados. Aunque menos exhaustiva que la búsqueda en rejilla, puede ser más eficiente en términos de tiempo y recursos.
Evaluación:

La selección de la mejor combinación de hiperparámetros se realiza utilizando un conjunto de validación o mediante validación cruzada, evaluando el rendimiento del modelo con diferentes métricas, como precisión, recall, y puntuación F1.
En resumen, la optimización de hiperparámetros es un proceso esencial para mejorar el rendimiento del modelo y lograr resultados más precisos y confiables en las tareas de Machine Learning.

## Redes Neuronales Convolucionales (CNNs)

---

Las Redes Neuronales Convolucionales (CNN, por sus siglas en inglés) son un tipo especializado de Redes Neuronales Artificiales diseñadas para procesar datos con estructura espacial, como imágenes. Inspiradas en el sistema visual biológico, las CNNs utilizan capas convolucionales para extraer características locales y patrones espaciales de los datos.

En términos simples, una CNN está compuesta por varias capas clave:

Capas Convolucionales: Estas capas aplican filtros (o kernels) a la entrada para detectar características locales como bordes, texturas y formas. Cada filtro se mueve a través de la imagen para generar un mapa de características, que resalta la presencia de ciertas características en distintas posiciones.

Capas de Activación: Después de la convolución, se aplica una función de activación, como ReLU (Rectified Linear Unit), para introducir no linealidades y permitir a la red aprender patrones complejos.

Capas de Pooling: También conocidas como capas de subsampling o reducción, estas capas reducen la dimensionalidad de los mapas de características, resumiendo la información y ayudando a que la red sea más eficiente y resistente a la variación en las posiciones de las características.

Capas Completamente Conectadas: Al final de la red, las capas completamente conectadas combinan las características extraídas para realizar una clasificación o regresión, dependiendo del tipo de tarea.

Las CNNs son altamente efectivas en el reconocimiento de patrones espaciales y son ampliamente utilizadas en aplicaciones como la clasificación de imágenes, el detection de objetos, y el reconocimiento facial.

En el caso de mi dataset, que está compuesto por información estructurada, exploraré el uso de CNNs para ver cómo se comportan en este tipo de datos, aunque están optimizadas principalmente para datos con estructura espacial como las imágenes

````python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score, classification_report, confusion_matrix
import seaborn as sns

# Definir la clase del modelo CNN
class CNNModel(nn.Module):
    def __init__(self, input_size, hidden_units, activation_fn):
        super(CNNModel, self).__init__()

        # Definir las capas convolucionales
        self.conv1 = nn.Conv1d(in_channels=1, out_channels=16, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv1d(in_channels=16, out_channels=32, kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv1d(in_channels=32, out_channels=64, kernel_size=3, stride=1, padding=1)

        # Definir la función de activación
        self.activation = activation_fn()

        # Definir un bloque de capas para simplificar el ajuste
        self.pool = nn.AdaptiveAvgPool1d(1)  # Reducción de la dimensión temporal

        # Definir las capas completamente conectadas
        self.fc1 = nn.Linear(64, hidden_units)
        self.fc2 = nn.Linear(hidden_units, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # Aplicar las capas convolucionales
        x = self.conv1(x)
        x = self.activation(x)
        x = self.conv2(x)
        x = self.activation(x)
        x = self.conv3(x)
        x = self.activation(x)

        # Aplicar el pooling adaptativo
        x = self.pool(x)

        # Aplanar la salida
        x = x.view(x.size(0), -1)

        # Aplicar las capas completamente conectadas
        x = self.fc1(x)
        x = self.activation(x)
        x = self.fc2(x)
        x = self.sigmoid(x)

        return x

# Separar características y variable objetivo
X = df_Smote.drop('Churn', axis=1)
y = df_Smote['Churn']

# Dividir el dataset en conjunto de entrenamiento y prueba
X_scaled = StandardScaler().fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Convertir a tensores
X_train_tensor = torch.tensor(X_train, dtype=torch.float32).unsqueeze(1)  # Añadir dimensión de canal
X_test_tensor = torch.tensor(X_test, dtype=torch.float32).unsqueeze(1)  # Añadir dimensión de canal
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).unsqueeze(1)  # Convertir a numpy array y añadir dimensión
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32).unsqueeze(1)  # Convertir a numpy array y añadir dimensión

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Parámetros a probar
hidden_units_options = [64, 128]
activation_options = {
    'ReLU': nn.ReLU,
    'Tanh': nn.Tanh
}

best_recall = 0
best_params = {}

# Búsqueda de hiperparámetros manual
for hidden_units in hidden_units_options:
    for activation_name, activation_fn in activation_options.items():
        model = CNNModel(input_size=X_train_tensor.shape[2], hidden_units=hidden_units, activation_fn=activation_fn)
        criterion = nn.BCELoss()
        optimizer = optim.Adam(model.parameters(), lr=0.001)

        # Inicializar listas para almacenar la pérdida
        train_losses = []

        # Entrenamiento del modelo
        model.train()
        for epoch in range(100):  # Número de épocas ajustado a 100
            epoch_losses = []
            for inputs, labels in train_loader:
                optimizer.zero_grad()
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                epoch_losses.append(loss.item())

            # Guardar la pérdida promedio para la época
            avg_loss = np.mean(epoch_losses)
            train_losses.append(avg_loss)

        # Evaluación del modelo
        model.eval()
        y_pred_probs = []
        y_true = []
        with torch.no_grad():
            for inputs, labels in test_loader:
                outputs = model(inputs)
                y_pred_probs.extend(outputs.numpy())
                y_true.extend(labels.numpy())

        y_pred = (np.array(y_pred_probs) > 0.5).astype(int).reshape(-1)
        y_true = np.array(y_true).astype(int).reshape(-1)

        recall = recall_score(y_true, y_pred)

        if recall > best_recall:
            best_recall = recall
            best_params = {
                'hidden_units': hidden_units,
                'activation_fn': activation_name,
                'train_losses': train_losses
            }

print(f"Mejores parámetros: {best_params}")
print(f"Mejor recall: {best_recall:.2f}")

# Evaluar el modelo con los mejores parámetros
best_activation_fn = activation_options[best_params['activation_fn']]
best_model = CNNModel(input_size=X_train_tensor.shape[2],
                      hidden_units=best_params['hidden_units'],
                      activation_fn=best_activation_fn)
criterion = nn.BCELoss()
optimizer = optim.Adam(best_model.parameters(), lr=0.001)
best_model.train()

# Inicializar lista para almacenar la pérdida
best_train_losses = best_params['train_losses']

for epoch in range(100):  # Número de épocas ajustado a 100
    epoch_losses = []
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = best_model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        epoch_losses.append(loss.item())

    # Guardar la pérdida promedio para la época
    avg_loss = np.mean(epoch_losses)
    best_train_losses.append(avg_loss)

best_model.eval()
y_pred_probs = []
y_true = []
with torch.no_grad():
    for inputs, labels in test_loader:
        outputs = best_model(inputs)
        y_pred_probs.extend(outputs.numpy())
        y_true.extend(labels.numpy())

y_pred = (np.array(y_pred_probs) > 0.5).astype(int).reshape(-1)
y_true = np.array(y_true).astype(int).reshape(-1)

# Evaluar el modelo
recall = recall_score(y_true, y_pred)
conf_matrix = confusion_matrix(y_true, y_pred)
class_report = classification_report(y_true, y_pred)

# Mostrar resultados
print(f'Recall del modelo: {recall:.2f}')
print('\nMatriz de Confusión:')
print(conf_matrix)
print('\nReporte de Clasificación:')
print(class_report)

# Visualizar la matriz de confusión
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Churn', 'Churn'], yticklabels=['No Churn', 'Churn'])
plt.title('Matriz de Confusión')
plt.xlabel('Predicción')
plt.ylabel('Realidad')
plt.show()

# Visualizar la evolución de la pérdida
plt.figure(figsize=(10, 6))
plt.plot(best_train_losses, label='Pérdida de Entrenamiento')
plt.xlabel('Época')
plt.ylabel('Pérdida')
plt.title('Evolución de la Pérdida durante el Entrenamiento')
plt.legend()
plt.xlim(0, 100)  # Ajustar el límite del eje x para que solo se vea hasta 100
plt.show()
````

## Redes Neuronales Recurrentes (RNNs)
---

Las Redes Neuronales Recurrentes (RNN, por sus siglas en inglés) son un tipo de red neuronal diseñada para trabajar con datos secuenciales o temporales, como series de tiempo o texto. A diferencia de las Redes Neuronales Artificiales (ANN) tradicionales y las Redes Neuronales Convolucionales (CNN), las RNN tienen la capacidad de mantener una memoria interna de las secuencias anteriores gracias a sus conexiones recurrentes. Esto significa que pueden tomar en cuenta información de pasos anteriores en la secuencia para hacer predicciones sobre pasos futuros, lo que las hace ideales para tareas donde el contexto temporal es crucial. Por ejemplo, en procesamiento de lenguaje natural, una RNN puede entender el significado de una palabra en una oración al considerar las palabras que la preceden.

En contraste, las ANN y las CNN están más orientadas a datos no secuenciales. Las ANN son redes neuronales básicas que suelen ser utilizadas para tareas generales de clasificación o regresión, donde las entradas no tienen un orden específico. Las CNN, por otro lado, están especializadas en el procesamiento de datos con estructura de rejilla, como imágenes, utilizando convoluciones para captar características espaciales. El mayor caso de uso de las RNN es el análisis de secuencias temporales, como en la predicción de series de tiempo, traducción automática, y el reconocimiento de voz, donde la capacidad de recordar y usar el contexto temporal es fundamental para la precisión del modelo.

````python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns

# Definir la clase del modelo RNN
class RNNModel(nn.Module):
    def __init__(self, input_size, hidden_units, activation_fn, num_layers=1):
        super(RNNModel, self).__init__()

        # Definir la capa RNN
        self.rnn = nn.RNN(input_size=input_size, hidden_size=hidden_units, num_layers=num_layers, batch_first=True)

        # Definir la función de activación
        self.activation = activation_fn()

        # Definir las capas completamente conectadas
        self.fc1 = nn.Linear(hidden_units, hidden_units)
        self.fc2 = nn.Linear(hidden_units, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # Aplicar la capa RNN
        h0 = torch.zeros(self.rnn.num_layers, x.size(0), self.rnn.hidden_size).to(x.device)
        out, _ = self.rnn(x, h0)

        # Usar la salida del último tiempo
        out = out[:, -1, :]

        # Aplicar las capas completamente conectadas
        out = self.fc1(out)
        out = self.activation(out)
        out = self.fc2(out)
        out = self.sigmoid(out)

        return out

# Separar características y variable objetivo
X = df_Smote.drop('Churn', axis=1)
y = df_Smote['Churn']

# Dividir el dataset en conjunto de entrenamiento y prueba
X_scaled = StandardScaler().fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Convertir a tensores y reestructurar
X_train_tensor = torch.tensor(X_train, dtype=torch.float32).unsqueeze(1)  # Añadir dimensión de secuencia
X_test_tensor = torch.tensor(X_test, dtype=torch.float32).unsqueeze(1)  # Añadir dimensión de secuencia
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).unsqueeze(1)  # Convertir a numpy array y añadir dimensión
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32).unsqueeze(1)  # Convertir a numpy array y añadir dimensión

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Parámetros a probar
hidden_units_options = [64, 128]
activation_options = {
    'ReLU': nn.ReLU,
    'Tanh': nn.Tanh
}

best_accuracy = 0
best_params = {}

# Búsqueda de hiperparámetros manual
for hidden_units in hidden_units_options:
    for activation_name, activation_fn in activation_options.items():
        model = RNNModel(input_size=X_train_tensor.shape[2], hidden_units=hidden_units, activation_fn=activation_fn)
        criterion = nn.BCELoss()
        optimizer = optim.Adam(model.parameters(), lr=0.001)

        # Inicializar listas para almacenar la pérdida
        train_losses = []

        # Entrenamiento del modelo
        model.train()
        for epoch in range(100):  # Número de épocas ajustado a 100
            epoch_losses = []
            for inputs, labels in train_loader:
                optimizer.zero_grad()
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                epoch_losses.append(loss.item())

            # Guardar la pérdida promedio para la época
            avg_loss = np.mean(epoch_losses)
            train_losses.append(avg_loss)

        # Evaluación del modelo
        model.eval()
        y_pred_probs = []
        y_true = []
        with torch.no_grad():
            for inputs, labels in test_loader:
                outputs = model(inputs)
                y_pred_probs.extend(outputs.numpy())
                y_true.extend(labels.numpy())

        y_pred = (np.array(y_pred_probs) > 0.5).astype(int).reshape(-1)
        y_true = np.array(y_true).astype(int).reshape(-1)

        accuracy = accuracy_score(y_true, y_pred)

        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_params = {
                'hidden_units': hidden_units,
                'activation_fn': activation_name,
                'train_losses': train_losses
            }

print(f"Mejores parámetros: {best_params}")
print(f"Mejor precisión: {best_accuracy:.2f}")

# Evaluar el modelo con los mejores parámetros
best_activation_fn = activation_options[best_params['activation_fn']]
best_model = RNNModel(input_size=X_train_tensor.shape[2],
                      hidden_units=best_params['hidden_units'],
                      activation_fn=best_activation_fn)
criterion = nn.BCELoss()
optimizer = optim.Adam(best_model.parameters(), lr=0.001)
best_model.train()

# Inicializar lista para almacenar la pérdida
best_train_losses = best_params['train_losses']

for epoch in range(100):  # Número de épocas ajustado a 100
    epoch_losses = []
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = best_model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        epoch_losses.append(loss.item())

    # Guardar la pérdida promedio para la época
    avg_loss = np.mean(epoch_losses)
    best_train_losses.append(avg_loss)

best_model.eval()
y_pred_probs = []
y_true = []
with torch.no_grad():
    for inputs, labels in test_loader:
        outputs = best_model(inputs)
        y_pred_probs.extend(outputs.numpy())
        y_true.extend(labels.numpy())

y_pred = (np.array(y_pred_probs) > 0.5).astype(int).reshape(-1)
y_true = np.array(y_true).astype(int).reshape(-1)

# Evaluar el modelo
accuracy = accuracy_score(y_true, y_pred)
conf_matrix = confusion_matrix(y_true, y_pred)
class_report = classification_report(y_true, y_pred)

# Mostrar resultados
print(f'Precisión del modelo: {accuracy:.2f}')
print('\nMatriz de Confusión:')
print(conf_matrix)
print('\nReporte de Clasificación:')
print(class_report)

# Visualizar la matriz de confusión
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Churn', 'Churn'], yticklabels=['No Churn', 'Churn'])
plt.title('Matriz de Confusión')
plt.xlabel('Predicción')
plt.ylabel('Realidad')
plt.show()

# Visualizar la evolución de la pérdida
plt.figure(figsize=(10, 6))
plt.plot(best_train_losses, label='Pérdida de Entrenamiento')
plt.xlabel('Época')
plt.ylabel('Pérdida')
plt.title('Evolución de la Pérdida durante el Entrenamiento')
plt.legend()
plt.xlim(0, 100)  # Ajustar el límite del eje x para que solo se vea hasta 100
plt.show()
````

Datos y sus Implicaciones
En la evaluación de modelos para la predicción de churn, es crucial comprender cómo se distribuyen los verdaderos positivos (VP), falsos negativos (FN), verdaderos negativos (VN) y falsos positivos (FP) en el conjunto de datos. Estos valores reflejan la capacidad del modelo para clasificar correctamente a los clientes en las categorías de interés, y cada tipo de error tiene implicaciones distintas para el negocio.

Verdaderos Positivos (VP) son aquellos clientes que realmente harán churn y son correctamente identificados como tales por el modelo. Un alto número de VP indica que el modelo es efectivo en captar a los clientes que están en riesgo de abandonar, lo cual es deseable en el contexto de un alto recall. El objetivo es maximizar este número para asegurar que la mayoría de los clientes que necesitan ser abordados sean identificados.

Falsos Negativos (FN) son clientes que realmente harán churn pero no son identificados como tales por el modelo. Un alto número de FN significa que el modelo está fallando en detectar a algunos clientes que podrían irse, lo que puede resultar en oportunidades perdidas para implementar estrategias de retención. Minimizar los FN es crucial para garantizar que no se pierdan clientes potenciales.

Verdaderos Negativos (VN) son clientes que no harán churn y son correctamente identificados como tales. Aunque no son el foco principal cuando se prioriza el recall, una cantidad adecuada de VN asegura que el modelo no esté clasificando incorrectamente a demasiados clientes como churn. Esto ayuda a reducir la tasa de falsos positivos.

Falsos Positivos (FP) son clientes que no harán churn pero son incorrectamente clasificados como churn por el modelo. Un alto número de FP puede llevar a esfuerzos y costos innecesarios para retener a clientes que en realidad no están en riesgo de abandonar. Este aspecto es especialmente importante para gestionar recursos de manera eficiente y evitar la sobrecarga en las estrategias de retención.

En resumen, mientras que un alto recall es vital para identificar a la mayoría de los clientes que harán churn, es igualmente importante equilibrar esta métrica con la tasa de falsos positivos para evitar costos adicionales y gestión ineficaz de recursos. El Random Forest ofrece el mejor balance entre recall y precisión en este caso, reduciendo los FP y maximizando los VP, lo cual lo hace una elección preferida para este escenario.
